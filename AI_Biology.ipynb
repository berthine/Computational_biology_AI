## Q1. Including the size and frequency of a, c, g,t and n in each sequence.

```python
## the fuction help to remove the nan:'N' and underscore _ in the sequence
## and get lenght of each sequence
def get_seq():
    dx = {}
    dx_seq ={}
    max_siz = 0
    for k in dict.keys():
        seq = dict[k]
        line = ''
        for char in seq:
            if char != '-'and char !='N':
                line = line+char
        if max_siz <=len(line):
            max_siz = len(line)
        dx[k]=len(line) #size of each sequence
        dx_seq[k]=line
    return dx,dx_seq  
```

```python
def one_mer(sequence, size=1):
    occurence = {}
    for x in range(len(sequence) - size + 1):
        letter_seq = sequence[x:x+size]
        if letter_seq in occurence:
            occurence[letter_seq] +=1
        else:
            occurence[letter_seq] = 1
        if 'N' not in occurence:
            occurence['N'] = 0
    return occurence
        
```

## Q2. Compute k-mer similarity matrix based on k from 3 to 7. Rank similarities against human.

```python
## k-mer function

def getKmers(sequence, size=1):
    return [sequence[x:x+size] for x in range(len(sequence) - size + 1)]
```

```python
## create a bag of word
def creatBog(seq,k):
    kmers = getKmers(seq, size=k)
    bog = {}
    for word in kmers:
        if word in bog.keys():
            bog[word] += 1
        else:
            bog[word] = 1
    return bog
```

```python
# get the occurence of each kmer
for seq in dx_seq:
    s = set((creatBog(dx_seq[seq],3).keys()))
```

```python
# co-ocurrence matrix will help to compute simularities 
def co_ocurance_mtrx(dx_seq,k):
    list_mers = []
    for seq in dx_seq:
        list_mers += (list(creatBog(dx_seq[seq],k).keys()))
        
    sets_mers = set(list_mers)
    coomatrx = []
    #print(sets_mers)
    for seq in dx_seq:
        row = []
        bog = creatBog(dx_seq[seq],k)
        for st in sets_mers:
            if st in bog.keys():
                row.append(bog[st])
            else:
                row.append(0)
        coomatrx.append(row)
    return np.array(coomatrx)
        
```

```python
from itertools import combinations
def get_combinations(arr):
    comb = combinations(arr,2)
    return list(comb)
```

```python
from numpy import dot
from numpy.linalg import norm

def get_simularities(co_mtx):
    row,col = np.shape(co_mtx)
    comb = get_combinations(np.arange(row))
    similarities = []
    names = list(dx_seq.keys())
    sim = {}
    for ind in comb:
        row1,row2 = ind
        a = co_mtx[row1][:]
        b = co_mtx[row2][:]
        cos_sim = dot(a, b)/(norm(a)*norm(b))
        similarities.append(cos_sim)
        name = names[row1] + ' and ' + names[row2]
        sim[name] = cos_sim
    return sim
```

```python
co_mtx = co_ocurance_mtrx(dx_seq,3)
similarities = get_simularities(co_mtx)
```
